Importing the dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score 
```

Data Collection and Data Preprocessing

```python
sonar_data = pd.read_csv(r'C:\Users\anany\OneDrive\Desktop\Machine Learning\Mine_vs_Rock_Prediction\Mine_vs_Rock_Prediction\sonar data.csv', header=None)
```

```python
sonar_data.head()
```

```python
#No of rows and columns
sonar_data.shape
```

```python
#Statistical data
sonar_data.describe()
```

```python
sonar_data[60].value_counts()
```

```python
sonar_data.groupby(60).mean()
```

```python
#Separating data and labels
X=sonar_data.drop(columns=60,axis=1)
Y=sonar_data[60]
```

```python
print(X)
print(Y)
```

Training and Test data

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.1,stratify=Y,random_state=1)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
print(X_train,Y_train)
```

Model Training - Logistic Regression

```python
#Training the logistic regression model with training data
model=LogisticRegression()
model.fit(X_train,Y_train)
```

Model Evaluation

```python
#Accuracy on Training data
X_train_predict=model.predict(X_train)
train_data_accuracy=accuracy_score(X_train_predict,Y_train)
```

```python
print('Accuracy of training data:',train_data_accuracy)
```

```python
#Accuracy on Testing data
X_test_predict=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_predict,Y_test)
```

```python
print('Accuracy of testing data:',test_data_accuracy)
```
